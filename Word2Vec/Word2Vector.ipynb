**Pre Processing the Data**
<br>
**Removing unnecessary special characters, additional spaces**
<br>
**Changing all characters to lowercase**

```python
#Replacing digits with a space in the paragraph
text= re.sub(r'\[[0-9]*\]', ' ', paragraph)
#Replacing whitespace, tabs, or newline with the single space.
text= re.sub(r'\s+', ' ', text)
#Lowering all the characters
text= text.lower()

text= re.sub(r'\d', ' ', text)
text= re.sub(r'\s+', ' ', text)
```

```python
text
```

```python
sentences= nltk.sent_tokenize(text)
sentences= [nltk.word_tokenize(sentence) for sentence in sentences]
#Remove the stopwords
for i in range(len(sentences)):
    sentences[i]= [word for word in sentences[i] if word not in stopwords.words('english')]
sentences
```

```python
##Training the Word2Vec Model
model= Word2Vec(sentences, min_count=1)
```

```python
words= list(model.wv.key_to_index.keys())
```

```python
words
```

```python
vector= model.wv['dream']
vector
```

```python
similar= model.wv.most_similar('faith')
similar
```
